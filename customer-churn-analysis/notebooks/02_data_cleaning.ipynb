## Data Cleaning

```python
## Import pandas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load raw data
data = pd.read_csv("../data/raw/Telco-Customer-Churn.csv")
data.head()
```

```python
## Info
data.info()
```

### Change datatype
TotalCharges column should be numeric instead of object.

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'].dtype
```

### Handle duplicate rows

```python
## Check duplicate
data.duplicated().sum()
```

```python
## Drop duplicate
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

### Handling Missing Values

```python
## Count missing values
data.isna().sum()/len(data)*100
```

#### Comment:
* There are 3 columns containing missing values.
* Two columns are numerical and one is categorical.
* Since the number of missing values is very low (less than 5%), we can drop those rows.

```python
## Drop missing rows
missing_cols = ['MonthlyCharges', 'TotalCharges', 'Dependents']
data.dropna(subset=missing_cols, inplace=True)
data.isna().sum()
```

### Drop unnecessary columns
* customerID column is not necessary for analysis, so we can drop it.

```python
data.drop(columns=['customerID'], inplace=True)
```

```python
data.columns
```

#### Comment:
Data is now clean and ready for analysis.

### Explort cleaned data

```python
data.to_csv('../data/processed/cleaned_telco_customer_churn.csv', index=False)
```
